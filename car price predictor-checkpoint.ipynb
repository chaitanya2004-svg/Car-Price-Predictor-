```python
import pandas as pd 
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression
from sklearn.linear_model import Lasso
from sklearn import metrics

```

```python
##loading the csv file to process on the data
```

```python
car_ds = pd.read_csv('CAR DETAILS FROM CAR DEKHO.csv')
```

```python
#now counting the seller type
print(car_ds.seller_type.value_counts())
```

```python
#now for transmission
print(car_ds.transmission.value_counts())
```

```python
#coverting the vlues of fuel type from string to numeric for better understandinh
car_ds.replace({'fuel':{'Diesel' : 0,'Petrol' : 1, 'CNG' : 2, 'LPG' : 3, 'Electric' : 4 }} , inplace = True)
               
#now for seller type
car_ds.replace({'seller_type':{'Individual' : 0,'Dealer' : 1, 'Trustmark Dealer' : 2}} , inplace = True)

#for transmission
car_ds.replace({'transmission':{'Manual' : 0,'Automatic' : 1}} , inplace = True)
#for owner
car_ds.replace({'owner':{'First Owner' : 1,'Second Owner' : 1, 'Third Owner' : 3 , 'Fourth & Above Owner' : 4 , 'Test Drive Car' : 5}} , inplace = True)
```

```python
car_ds.infer_objects(copy=False)
```

```python
# Set pandas option to opt into the future behavior
pd.set_option('future.no_silent_downcasting', True)
```

```python
car_ds.head()
```

```python
#spliiting the data into training data and the test data
X = car_ds.drop(['selling_price', 'name'], axis = 1)
Y = car_ds['selling_price']
                
```

```python
print(X)
```

```python
print(Y)
```

```python
X_train, X_test, Y_train, Y_test = train_test_split(X, Y, test_size=0.2, random_state=42)
```

```python
#Linear Regression model
lin_reg_model = LinearRegression()
```

```python
#fit the data into linear regression
lin_reg_model.fit(X_train, Y_train)
```

```python
#prediction on the training data
training_data_pred = lin_reg_model.predict(X_train)
```

```python
#r square error 
error_score = metrics.r2_score( Y_train,training_data_pred)
print("R square error :" ,error_score)
```

```python
#plotting the values on the graph
plt.scatter( Y_train,training_data_pred)
plt.xlabel("Actual Price")
plt.ylabel("Predited Price")
plt.title("Actual Price vs Predited Price")
plt.show()
```

```python
#prediction on the test data
test_data_pred = lin_reg_model.predict(X_test)
```

```python
error_score = metrics.r2_score( Y_test,test_data_pred)
print("R square error :" ,error_score)
```

```python
plt.scatter( Y_test,test_data_pred)
plt.xlabel("Actual Price")
plt.ylabel("Predited Price")
plt.title("Actual Price vs Predited Price")
plt.show()
```
